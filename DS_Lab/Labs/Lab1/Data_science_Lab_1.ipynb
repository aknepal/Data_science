# **Computation of measures of Central Tendency**

Mean

```python
mean = df_csv['sepal_length'].mean()
print(mean)
```

Median

```python
median = df_csv['sepal_length'].median()
print(median)
```

Mode

```python
mode = df_csv['sepal_length'].mode()
print(mode)
```

# **Observation**

Here, mean > median > mode.

# **Plotting Distribuition**

```python
sns.displot(df_csv['sepal_length'], kde=True, bins=30, kind="hist")
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

# **Computation of measures of dispersion or variability**

Minimum value

```python
df_csv['sepal_length'].min()
```

Maximum value

```python
df_csv['sepal_length'].max()
```

Range

```python
df_csv['sepal_length'].max() - df_csv['sepal_length'].min()
```

Variance

```python
df_csv['sepal_length'].var()
```

Standard Deviation

```python
df_csv['sepal_length'].std()
```

First Quartile

```python
Q1= df_csv['sepal_length'].quantile(0.25)
Q1
```

Second Quartile

```python
Q2 = df_csv['sepal_length'].quantile(0.5)
Q2
```

Third Quartile

```python
Q3= df_csv['sepal_length'].quantile(0.75)
Q3
```

Inter Quartile Range

```python
IQR = Q3 -Q1
IQR
```
